```python
import pandas as pd
import os

def excel_to_txt(input_excel_file, output_txt_folder):
    # Đọc file Excel
    df = pd.read_excel(input_excel_file)

    # Kiểm tra xem "name" có trong các cột hay không
    if 'Tên' not in df.columns:
        print("Không tìm thấy cột 'Tên' trong file Excel.")
        return

    # Lặp qua từng dòng và ghép thành nội dung text
    for index, row in df.iterrows():
        text_content = '\n'.join(str(row[col]) for col in df.columns if col != 'Điều Khoản' and pd.notna(row[col])) 
        name_column_value = str(row['Tên']).replace(" ", "_")  # Thay thế khoảng trắng bằng dấu gạch dưới
        base_output_file_path = f"{output_txt_folder}/{name_column_value}.txt"
        
        # Kiểm tra xem file đã tồn tại hay chưa
        output_file_path = base_output_file_path
        counter = 1
        while os.path.exists(output_file_path):
            output_file_path = f"{output_txt_folder}/luatdt22_{name_column_value}_{counter}.txt"
            counter += 1

        # Lưu nội dung text vào file txt
        with open(output_file_path, 'w', encoding='utf-8') as file:
            file.write(text_content)

        print(f"Đã lưu nội dung vào file: {output_file_path}")
        
def excel_to_txt_update(input_excel_file, output_txt_folder, text_type, ten_vanban):
    df = pd.read_excel(input_excel_file)
    
    for index, row in df.iterrows():
        # Kiểm tra giá trị của cột Note
        note_value = row['Note']

        # Xây dựng nội dung cho file .txt dựa trên điều kiện
        if pd.isnull(note_value) or note_value == 0:
            content = ten_vanban + "\n".join([str(row[col]) for col in df.columns if col not in ["Chương", "Tiêu đề Chương", "Tên Mục", 'content', 'Note'] and pd.notna(row[col])])
            # Tạo tên file .txt dựa trên giá trị của cột Chương và Tiêu đề Chương
            name_column_value = str(row['Tên']).replace(" ", "_")  # Thay thế khoảng trắng bằng dấu gạch dưới
            file_name = f"{text_type}_{name_column_value}.txt"
            # Đường dẫn đầy đủ cho file .txt
            txt_file_path = os.path.join(output_txt_folder, file_name)

            # Kiểm tra xem file đã tồn tại hay chưa
            output_file_path = txt_file_path
            counter = 1
            while os.path.exists(output_file_path):
                counter += 1
                output_file_path = f"{output_txt_folder}/{text_type}_{name_column_value}_{counter}.txt"

            # Lưu nội dung text vào file txt
            with open(output_file_path, 'w', encoding='utf-8') as file:
                file.write(content)
            print(f"Đã lưu nội dung vào file: {output_file_path}")
        elif note_value == 1:
            content = ten_vanban + "\n".join([str(row[col]) for col in df.columns if col not in ["Chương", "Tiêu đề Chương", "Tên Mục",'Điều Khoản', 'Note'] and pd.notna(row[col])])

            # Tạo tên file .txt dựa trên giá trị của cột Chương và Tiêu đề Chương
            name_column_value = str(row['Tên']).replace(" ", "_")  # Thay thế khoảng trắng bằng dấu gạch dưới
            file_name = f"{text_type}_{name_column_value}.txt"
            # Đường dẫn đầy đủ cho file .txt
            txt_file_path = os.path.join(output_txt_folder, file_name)

            # Kiểm tra xem file đã tồn tại hay chưa
            output_file_path = txt_file_path
            # Lưu nội dung text vào file txt
            with open(output_file_path, 'w', encoding='utf-8') as file:
                file.write(content)
            print(f"Đã lưu nội dung vào file: {output_file_path}")
```

```python
import os
import random
import shutil

def split_data(input_folder, train_folder, val_folder, split_ratio=0.8, seed=None):
    # Kiểm tra xem tỷ lệ chia có hợp lý hay không
    if split_ratio < 0 or split_ratio > 1:
        print("Tỷ lệ chia không hợp lý. Tỷ lệ chia nên nằm trong khoảng từ 0 đến 1.")
        return

    # Thiết lập seed cho việc tạo số ngẫu nhiên để có thể tái tạo kết quả
    random.seed(seed)

    # Tạo thư mục train và test nếu chưa tồn tại
    os.makedirs(train_folder, exist_ok=True)
    os.makedirs(val_folder, exist_ok=True)

    # Lấy danh sách tất cả các file trong thư mục input
    all_files = [f for f in os.listdir(input_folder) if os.path.isfile(os.path.join(input_folder, f))]

    # Số lượng file cần chia vào thư mục train
    train_count = int(len(all_files) * split_ratio)

    # Chọn ngẫu nhiên các file cho thư mục train
    train_files = random.sample(all_files, train_count)

    # Di chuyển các file vào thư mục train
    for file_name in train_files:
        source_path = os.path.join(input_folder, file_name)
        destination_path = os.path.join(train_folder, file_name)
        shutil.copy(source_path, destination_path)

    # Lấy danh sách các file còn lại
    remaining_files = list(set(all_files) - set(train_files))

    # Số lượng file cần chia vào thư mục val
    val_count = len(all_files) - train_count

    # Chọn ngẫu nhiên các file cho thư mục val
    val_files = random.sample(remaining_files, val_count)

    # Copy các file vào thư mục val
    for file_name in val_files:
        source_path = os.path.join(input_folder, file_name)
        destination_path = os.path.join(val_folder, file_name)
        shutil.copy(source_path, destination_path)

input_folder_path = '../Raw_data/vibilaw/bilaw22_text_data'
train_folder_path = '../Raw_data/vibilaw/training_data/train'
val_folder_path = '../Raw_data/vibilaw/training_data/val'

# Tỉ lệ chia là 80-20%
split_ratio = 0.9

split_data(input_folder_path, train_folder_path, val_folder_path, split_ratio)
```
